```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
def sin_taylor(x, n_terms):
    approximation = 0
    for n in range(n_terms):
        approximation += ((-1)**n * x**(2*n + 1)) / factorial(2*n + 1)
    return approximation
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = np.linspace(-2 * math.pi, 2 * math.pi, 400)
```

```python
sc_x = sc.fit_transform(x.reshape(-1, 1))
```

```python
y_approx = [sin_taylor(xi, 5) for xi in x]
```

```python
y_approx_sc = [sin_taylor(xi, 5) for xi in sc_x]
```

```python
y_real = np.sin(x)
```

```python
y_real_sc = np.sin(sc_x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y_real, label='실제 sin(x)', color='blue')
plt.plot(x, y_approx, label='테일러 급수 근사치 (5항)', color='red', linestyle='dashed')
plt.title("실제 sin(x) 값 vs 테일러 급수 근사치")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sc_x, y_real_sc, label='실제 sin(x)', color='blue')
plt.plot(sc_x, y_approx_sc, label='테일러 급수 근사치 (5항)', color='red', linestyle='dashed')
plt.title("실제 sin(x) 값 vs 테일러 급수 근사치")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 400).reshape(-1, 1)

# y = x^2 함수로 y값을 생성
y = x**2

# x만 스케일링
scaler_x = StandardScaler()
x_scaled = scaler_x.fit_transform(x)

# x 스케일 기준으로 y도 스케일링
y_scaled_xbasis = (y - scaler_x.mean_[0]) / np.sqrt(scaler_x.var_[0])

# x와 y를 각각 독립적으로 스케일링
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y)

# 세 가지 경우로 그래프를 그림
plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.plot(x, y, label='원래 함수: y = x^2')
plt.title("원래 함수")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x_scaled, y_scaled_xbasis, label='x 기준으로 y도 스케일링', color='blue')
plt.title("x 기준으로 y도 스케일링한 함수")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled_xbasis")
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x_scaled, y_scaled, label='x, y 각각 스케일링', color='red')
plt.title("x, y 각각 스케일링한 함수")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
