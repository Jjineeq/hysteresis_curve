```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
def sin_taylor(x, n_terms):
    approximation = 0
    for n in range(n_terms):
        approximation += ((-1)**n * x**(2*n + 1)) / factorial(2*n + 1)
    return approximation
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = np.linspace(-2 * math.pi, 2 * math.pi, 400)
```

```python
sc_x = sc.fit_transform(x.reshape(-1, 1))
```

```python
y_approx = [sin_taylor(xi, 5) for xi in x]
```

```python
y_approx_sc = [sin_taylor(xi, 5) for xi in sc_x]
```

```python
y_real = np.sin(x)
```

```python
y_real_sc = np.sin(sc_x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y_real, label=' sin(x)', color='blue')
plt.plot(x, y_approx, label='   (5)', color='red', linestyle='dashed')
plt.title(" sin(x)  vs   ")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sc_x, y_real_sc, label=' sin(x)', color='blue')
plt.plot(sc_x, y_approx_sc, label='   (5)', color='red', linestyle='dashed')
plt.title(" sin(x)  vs   ")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 400).reshape(-1, 1)

# y = x^2  y 
y = x**2

# x 
scaler_x = StandardScaler()
x_scaled = scaler_x.fit_transform(x)

# x   y 
y_scaled_xbasis = (y - scaler_x.mean_[0]) / np.sqrt(scaler_x.var_[0])

# x y   
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y)

#     
plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.plot(x, y, label=' : y = x^2')
plt.title(" ")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x_scaled, y_scaled_xbasis, label='x  y ', color='blue')
plt.title("x  y  ")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled_xbasis")
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x_scaled, y_scaled, label='x, y  ', color='red')
plt.title("x, y   ")
plt.xlabel("x_scaled")
plt.ylabel("y_scaled")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
