```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# 데이터 생성
x = np.linspace(0, 2*np.pi, 100)
y = np.linspace(0, 2*np.pi, 100)
X, Y = np.meshgrid(x, y)

k = 2.0
F = np.sin(X) * np.sin(Y)
U = np.sin(k * X) * np.sin(k * Y)
```

```python
import matplotlib.pyplot as plt
```

```python
input_data = np.stack([X.ravel(), Y.ravel(), F.ravel()], axis=1)
output_data = U.reshape(-1, 1)

# Neural Operator 모델 정의
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(3,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```

```python
# 학습
his = model.fit(input_data, output_data, epochs=100, batch_size=32)

# 예측
U_pred = model.predict(input_data)
```

```python
plt.plot(his.history['loss'])
```
